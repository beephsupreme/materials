```python
import pandas as pd
import numpy as np
```

```python
materials = pd.read_excel('C:\\Users\\beeph\\projects\\jupyter\\materials\\data\\materials.xlsx')
bom = pd.read_csv('C:\\Users\\beeph\\projects\\jupyter\\materials\\data\\bom.txt')
bom = bom.drop('UM_Multiplier', axis=1)
bom = bom.groupby('Part Number', as_index=False).sum()
dates = list(materials.columns[9:])
cols_to_drop = ['T-Avail', 'R-Avail', 'Reorder']
cols_to_drop = cols_to_drop + dates
materials = materials.merge(bom, how='right').fillna(0)
materials = materials.drop(cols_to_drop, axis=1)
materials['Available'] = materials['On Hand'] - materials['Qty Ordered'] - materials['Backlog']
idx = materials[materials['Part Number'] == " "].index
materials = materials.drop(idx, axis=0)
```

```python
def check_it(on_hand, available):
    if available < 1:
        return "NFB"
    ratio = available / on_hand
    if ratio < .1:
        return "NFB"
    elif ratio < .25:
        return "CAUTION"
    else:
        return "OK"
    
```

```python
materials['Status'] = materials[['On Hand','Available']].apply(lambda materials: check_it(materials['On Hand'], materials['Available']), axis=1)
```
